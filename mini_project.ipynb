```python
import pandas as pd
import numpy as np
spotifyTop = pd.read_csv("/content/final.csv")
spotifyTop.head(10)
```

```python
for col in (spotifyTop.columns):
    print(spotifyTop[col].describe())
    print("")
    print("")
```

```python
print(spotifyTop.columns)
```

```python
dataOfInterest = spotifyTop[['track_name', 'rank', 'danceability', 'energy', 'key', 'mode', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration', 'weeks_on_chart', "streams"]]
dataOfInterest.head(5)
```

```python
# convert data to numbers
def convertToNum(dataOfInterest):
    for colNames in dataOfInterest.columns:
        if colNames == 'track_name':
            continue
        dataOfInterest[colNames] = pd.to_numeric(dataOfInterest[colNames], errors='coerce')

convertToNum(dataOfInterest)
dataOfInterest.head(5)
```

```python
dataOfInterest = dataOfInterest.groupby('track_name').agg({
      'rank': 'min',
      'weeks_on_chart': 'max',
      'streams': 'max',
      'danceability': 'mean',
      'energy': 'mean',
      'key':'mean',
      'mode':'mean',
      'loudness':'mean',
      'speechiness':'mean',
      'acousticness':'mean',
      'instrumentalness':'mean',
      'liveness':'mean',
      'valence':'mean',
      'tempo':'mean',
      'duration':'mean'
})
dataOfInterest.head(5)
```

```python
dataOfInterest["binning"] = pd.cut(dataOfInterest["rank"], [0, 100,  200], labels=['0-100', '100-200'])
pd.unique(dataOfInterest["binning"])
```

```python
for colNames in dataOfInterest.columns:
    if colNames == "binning" or colNames == 'track_name':
        continue;
    dataOfInterest[colNames] = (dataOfInterest[colNames] - dataOfInterest[colNames].min()) / (dataOfInterest[colNames].max() - dataOfInterest[colNames].min())
dataOfInterest.describe()
```

```python
def printNaDetails(dataOfInterest):
    for colNames in dataOfInterest.columns:
        total = len(dataOfInterest[colNames])
        numberOfNa = dataOfInterest[colNames].isna().sum()
        perOfNa = (numberOfNa/total)*100
        print(colNames, end=' ')
        print(total, end=', ')
        print(numberOfNa, end=', ')
        print(perOfNa, end="\n")

printNaDetails(dataOfInterest)
```

```python
dataOfInterest = dataOfInterest.dropna()
printNaDetails(dataOfInterest)
```

```python
bins = dataOfInterest.groupby('binning')
bin1 = bins.get_group('0-100')
bin2 = bins.get_group('100-200')
import matplotlib.pyplot as plt
x = ['0-100', '100-200']
y = [len(bin1), len(bin2)]
fig, ax = plt.subplots()

p1 = ax.bar(x, y, color = "maroon")

ax.set_ylabel("Number of values")
ax.set_xlabel("Ranges")
ax.set_title("Number of values vs Ranges")
# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='edge')

plt.show()
```

```python
corr = []
for col in ['danceability', 'energy', 'key', 'mode', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration']:
    corr.append(np.corrcoef(dataOfInterest[col], dataOfInterest["weeks_on_chart"])[0,1])


dictionary = {}

dictionary['attributes'] = ['danceability', 'energy', 'key', 'mode', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration']
dictionary['correlation'] = corr
pd.DataFrame(dictionary)
```

```python
from sklearn.model_selection import train_test_split

SEED = 42
X = dataOfInterest[['danceability', 'energy', 'key', 'mode', 'loudness',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration']]
y = dataOfInterest['binning']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED)
from sklearn.preprocessing import LabelEncoder

# Creating a instance of label Encoder.
le = LabelEncoder()

# Using .fit_transform function to fit label
# encoder and return encoded label
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
print(pd.unique(y_train))
print(pd.unique(y_test))
# for col in (X_test.columns):
#     print(X_test[col].describe())
#     print("")
#     print("")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

error = []
accuracy = []

# Calculating MAE error for K values between 1 and 39
for i in range(1, 100):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    score = metrics.accuracy_score(y_test, pred_i)
    accuracy.append(score)
    # Model Accuracy, how often is the classifier correct?
    print("Accuracy:", score)
```

```python
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train, y_train)
pred_i = knn.predict(X_test)
score = metrics.accuracy_score(y_test, pred_i)
# Model Accuracy, how often is the classifier correct?
print("Accuracy:", score)
```
